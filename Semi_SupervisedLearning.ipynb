```python
import numpy as np
from sklearn.semi_supervised import LabelPropagation
import matplotlib.pyplot as plt
```

```python
#Create synthetic data
X = np.array([[1, 1], [2, 1], [3, 2], [5, 5], [6, 5], [7, 6]])
y = np.array([0, 0, 1, -1, -1, -1]) # -1 represents unlabeled data
```

```python
#Initialize label Propagation model
model = LabelPropagation()

model.fit(X,y)
```

```python
#Predict labels for unlabeled data

predicted_label = model.transduction_

print(f"Predicted labels: {predicted_label}")
```

```python
#Plotting the data

plt.scatter(X[:,0],X[:,1],c=predicted_label,cmap="viridis", s=100)
plt.show()
```
